```python
import pandas as pd 
```

```python
df=pd.DataFrame({"Size":["s","m","l","xl","s","xl"]})
```

```python
df
```

```python
ord_df=[["s","m","l","xl"]]
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe = OrdinalEncoder(categories=ord_df)
```

```python
oe.fit(df[["Size"]])
```

```python
df["Size_en"]=oe.transform(df[["Size"]])
```
